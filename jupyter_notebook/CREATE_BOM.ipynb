```python
import pandas as pd

pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 40)
sx = pd.read_excel('DATA TINH BOM.xls', sheet_name='SX')
nvl = pd.read_excel('DATA TINH BOM.xls', sheet_name='NVL')
# DELETE IN PRACTICE
sx.drop_duplicates(inplace=True)
result = pd.merge(nvl, sx, how="left", on=['Manufacturing Order No'], sort=False,
                              indicator=False, validate=None)
result['BOM'] = result['Issue Quantity'] / result['Prod. Qty.']
```
